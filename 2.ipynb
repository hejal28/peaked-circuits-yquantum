```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
import numpy as np

# Load circuit
qc = QuantumCircuit.from_qasm_file("P2_small_bump (1).qasm")

print("Qubits:", qc.num_qubits)

# Use statevector (safe for ~12 qubits)
sim = AerSimulator(method="statevector")

qc.save_statevector()

result = sim.run(qc).result()
statevector = result.get_statevector()

# Compute probabilities
probs = np.abs(statevector)**2

# Find peak
peak_index = np.argmax(probs)

# Convert to bitstring
peak = format(peak_index, f"0{qc.num_qubits}b")

print("Peak bitstring:", peak)
```
