```python
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator

qc = QuantumCircuit.from_qasm_file("P4_gentle_mound (1).qasm")

print("Qubits:", qc.num_qubits)
print("Depth:", qc.depth())

qc.measure_all()

sim = AerSimulator(
    method="matrix_product_state",
    matrix_product_state_max_bond_dimension=64
)

# slightly more shots for safety
result = sim.run(qc, shots=800).result()

counts = result.get_counts()

# top candidates
top_states = sorted(counts.items(), key=lambda x: x[1], reverse=True)[:3]

print("Top candidates:")
for s, c in top_states:
    print(s, c)

best = top_states[0][0]

print("\nTry these:")
print("1) Raw      :", best)
print("2) Reversed :", best[::-1])
```
