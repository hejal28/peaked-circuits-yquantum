```python
from qiskit import QuantumCircuit
import numpy as np

qc = QuantumCircuit.from_qasm_file("P1_little_peak (2).qasm")

bits = ["0"] * qc.num_qubits

for instr in qc.data:
    op = instr.operation
    q = instr.qubits[0]._index

    if op.name == "x":
        bits[q] = "1" if bits[q] == "0" else "0"

    elif op.name == "ry":
        theta = float(op.params[0])

        p0 = np.cos(theta/2)**2
        p1 = np.sin(theta/2)**2

        bits[q] = "1" if p1 > p0 else "0"

peak = "".join(bits[::-1])

print("Peak bitstring:", peak)
```
