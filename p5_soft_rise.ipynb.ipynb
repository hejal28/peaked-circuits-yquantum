```python
import bluequbit
from qiskit import QuantumCircuit
import time

# 🔑 your token
bq = bluequbit.init("")

qc = QuantumCircuit.from_qasm_file("P5_soft_rise (1).qasm")
n = qc.num_qubits

def get_pauli(i, n):
    return "I"*i + "Z" + "I"*(n-i-1)

bits = []
start_total = time.time()

for i in range(n):
    start = time.time()
    
    print(f"\n[{i+1}/{n}] Solving qubit {i}...")

    pauli = [(get_pauli(i, n), 1.0)]
    
    job = bq.run(
        qc,
        device="pauli-path",
        pauli_sum=pauli,
        options={
            "pauli_path_truncation_threshold": 1e-4
        }
    )
    
    val = job.expectation_value
    bit = '1' if val < 0 else '0'
    
    bits.append(bit)
    
    elapsed = time.time() - start
    total_elapsed = time.time() - start_total
    
    progress = (i+1)/n * 100
    partial = "".join(bits[::-1])  # current reconstructed string
    
    print(f"  ⟨Z⟩ = {val:.4f} → {bit}")
    print(f"  Time: {elapsed:.2f}s | Progress: {progress:.1f}%")
    print(f"  Partial (reversed): {partial}")
    print(f"  Total elapsed: {total_elapsed:.1f}s")

# final result
final = "".join(bits[::-1])

print("\n" + "="*60)
print("FINAL ANSWER:")
print(final)
print("="*60)
```
